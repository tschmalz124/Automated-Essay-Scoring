```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
single_scaler = pd.read_csv('data\interim\single_scaler_results.csv', index_col=0)
essay_dependent_scaler = pd.read_csv('data\interim\essay_dependent_scaler_results.csv', index_col=0)
```

```python
single_scaler['residuals_scaled'] = single_scaler.orig_scaled - single_scaler.pred_scaled
single_scaler['residuals_unscaled'] = single_scaler.orig_unscaled - single_scaler.pred_unscaled
```

```python
essay_dependent_scaler['residuals_scaled'] = essay_dependent_scaler.orig_scaled - essay_dependent_scaler.pred_scaled
essay_dependent_scaler['residuals_unscaled'] = essay_dependent_scaler.orig_unscaled - essay_dependent_scaler.pred_unscaled
```

```python
sns.pairplot(data=single_scaler, hue='essay_set')
plt.savefig('figures\single_scaler_pairplot.png')
```

```python
sns.pairplot(data=essay_dependent_scaler, hue='essay_set')
plt.savefig('figures\essay_dependent_scaler_pairplot.png')
```

### Single Scaler

```python
sns.lmplot(x='pred_unscaled', y='orig_unscaled', data=single_scaler, hue='essay_set', fit_reg = False)
plt.title('Single scaler results')
```

```python
sns.lmplot(x='pred_scaled', y='orig_scaled', data=single_scaler, hue='essay_set', fit_reg = False)
plt.title('Single scaler results')
```

```python
sns.lmplot(x='orig_unscaled', y='residuals_unscaled', data=single_scaler, hue = 'essay_set', fit_reg=False)
```

```python
sns.lmplot(x='orig_scaled', y='residuals_scaled', data=single_scaler, hue = 'essay_set', fit_reg=False)
```

```python
sns.boxplot(x='essay_set', y='residuals_scaled', data=single_scaler)
```
